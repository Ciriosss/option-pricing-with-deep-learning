#### EXTRACTING DATA

## Extracting ticker

```python
oiginal_data = pd.read_csv("Data/Full_data.csv")
tickers_symbol = oiginal_data.ticker.unique().tolist()
```

```python
import yfinance as yf
import pandas as pd
from datetime import datetime

rnn_data_to_add = pd.DataFrame()


for ticker_symbol in tickers_symbol:
    try:
        ticker = yf.Ticker(ticker_symbol)
        option_chain = ticker.option_chain()
        try:
            ticker_calls = option_chain.calls  # Extract calls data
            ticker_calls['ticker'] = ticker_symbol  # Add a column for ticker symbol
            ticker_calls['call'] = 1 # Add a column for ticker symbol
            rnn_data_to_add = pd.concat([rnn_data_to_add, ticker_calls])  # Concatenate calls data to the DataFrame
        except:
            pass
        
        try:
            ticker_puts = option_chain.puts  # Extract puts data
            ticker_puts['ticker'] = ticker_symbol  # Add a column for ticker symbol
            ticker_puts['call'] = 0 # Add a column for ticker symbol
            rnn_data_to_add = pd.concat([rnn_data_to_add, ticker_puts])  # Concatenate puts data to the DataFrame
        except:
            pass
    except:
        pass
    
rnn_data_to_add
```

```python
from datetime import date

today = date.today() 


rnn_data_to_add['date'] = today
rnn_data_to_add
```

```python

rnn_data.drop("Unnamed: 0",axis = 1, inplace = True)
```

```python
from datetime import datetime, timedelta
import yfinance as yf

def extract_data(lastTradeDate):
    data_datetime = datetime.strptime(str(lastTradeDate), "%Y-%m-%d %H:%M:%S%z")
    return data_datetime.date()

# Apply function to each row
rnn_data_to_add['lastTradeDate'] = rnn_data_to_add['lastTradeDate'].apply(extract_data)
```

```python

rnn_data_to_add.drop('currency', axis=1, inplace=True)
```

```python
def stock_price(row):

    trade_date =  datetime.strptime(str(row['lastTradeDate']), "%Y-%m-%d")
    end = trade_date +  timedelta(days=3)
    start = trade_date - timedelta(days=3)
    ticker = row['ticker']
    stock = yf.Ticker(ticker)


    historical_data = stock.history(start=start, end=end)


    def get_stock_price(date, data):
        try:
                return data.loc[date.strftime('%Y-%m-%d')]['Close']
        except KeyError:
                return None

    stock_price = None
    for i in range(4):
        if stock_price is None:
                stock_price = get_stock_price(trade_date - timedelta(days=i), historical_data)
        if stock_price is None:
                stock_price = get_stock_price(trade_date + timedelta(days=i), historical_data)

        
    return stock_price


rnn_data_to_add['stock'] = rnn_data_to_add.apply(stock_price, axis=1)
rnn_data_to_add = rnn_data_to_add[rnn_data_to_add['stock'] != 'False']
rnn_data_to_add['stock'] = rnn_data_to_add['stock'].astype('float')
```

```python
rnn_data_to_add
```

```python

rnn_data = rnn_data[rnn_data.stock > 0]
```

```python
rnn_data = pd.concat([rnn_data,rnn_data_to_add])
```
