```python
import pandas as pd 

data = pd.read_csv('Data/full_rnn_data_2.csv')
data.drop('Unnamed: 0', axis = 1, inplace= True)
data.drop('state', axis = 1, inplace= True)
data
```

```python
data.recommendationKey.unique()
```

```python
missing_percentage = (data.isnull().sum() / len(data)) * 100
missing = missing_percentage[(missing_percentage > 0)]
to_fill_median = missing.index.tolist()

# List of columns to remove
columns_to_remove = ['stock', 'address1', 'city', 'country', 'industry', 'sector','recommendationKey']

# Remove each column from the list
for col in columns_to_remove:
    if col in to_fill_median:  # Check if the column is in the list
        to_fill_median.remove(col)

to_fill_median
```

```python
columns_to_remove.remove('stock')
```

```python
columns_to_remove
```

```python
column_median = data[to_fill_median].median()
data[to_fill_median] = data[to_fill_median].fillna(column_median)
```

```python
data = data.drop(columns=columns_to_remove)
```

```python
missing_percentage = (data.isnull().sum() / len(data)) * 100
missing = missing_percentage[(missing_percentage > 0)]
missing
```

```python
data.to_csv('Data/full_rnn_data_2.csv')
```
